# mirrord Log Analysis
Notebook for post-mirrord trace inspection.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

trace_path = Path('ops/mirrord/trace-report.json')
if not trace_path.exists():
    raise FileNotFoundError(f'Missing trace file: {trace_path}')

with trace_path.open() as f:
    data = json.load(f)

df = pd.DataFrame(data.get('events', []))
if df.empty:
    raise ValueError('Trace report contains no events to analyze.')

print('Average latency (ms):', df['latency_ms'].mean())
print('Max latency (ms):', df['latency_ms'].max())

plt.figure(figsize=(10, 4))
plt.plot(pd.to_datetime(df['timestamp']), df['latency_ms'], marker='o')
plt.title('Mirrord Network Latency Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Latency (ms)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
